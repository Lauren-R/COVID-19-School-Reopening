## 02. Clean data

```python
# Find duplicates

data_dups = data[data.duplicated()]
```

```python
data_dups.shape
```

##### No duplicates

```python
# Check for columns with mixed data types

for col in data.columns.tolist():
      weird = (data[[col]].applymap(type) != data[[col]].iloc[0].apply(type)).any(axis = 1)
      if len (data[weird]) > 0:
        print (col)
```
