## Create a correlation matrix heatmap (colored)

```python
# Create a correlation heatmap using matplotlib

plt.matshow(group1.corr())
plt.show()
```

```python
# Add labels, a legend, and change the size of the heatmap

f = plt.figure(figsize=(12, 12)) # figure size 
plt.matshow(group1.corr(), fignum=f.number) # type of plot
plt.xticks(range(group1.shape[1]), group1.columns, fontsize=10, rotation=90) # x axis labels
plt.yticks(range(group1.shape[1]), group1.columns, fontsize=10) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=14) # add font size
plt.title('Correlation Matrix Heatmap', fontsize=14) # add title
```
