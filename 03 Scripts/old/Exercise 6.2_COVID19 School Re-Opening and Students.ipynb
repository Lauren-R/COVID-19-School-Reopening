## 02. Setup variables for analysis

```python
# Create new column - ELL and migrant students

df_merged['ELL  migrant students'] = df_merged['English Language Learners'] + df_merged['Migrant'] / df_merged['All Students']
```

```python
df_merged.head(50)
```

```python
# Create new column - homeless and low-income students

df_merged['homeless and low-income students'] = df_merged['Homeless'] + df_merged['Low-Income'] / df_merged['All Students']
```

```python
# Create new column - section 504 and students with disabilities

df_merged['section 504 and students with disabilities'] = df_merged['Section 504'] + df_merged['Students with Disabilities'] / df_merged['All Students']
```
