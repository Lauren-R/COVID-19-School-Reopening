## 02. Clean data

```python
# Find missing values - reopening

reopening.isnull().sum()

# Columns with missing values are expected due to nature of variables.
```

```python
# reopening df - Check for columns with mixed data types

for col in reopening.columns.tolist():
      weird = (reopening[[col]].applymap(type) != reopening[[col]].iloc[0].apply(type)).any(axis = 1)
      if len (reopening[weird]) > 0:
        print (col)
```

```python
# reopening df - Change mixed data type columns to string

reopening['Status'] = reopening['Status'].astype('str')
reopening['Model_Elementary'] = reopening['Model_Elementary'].astype('str') 
reopening['Model_Middle'] = reopening['Model_Middle'].astype('str')
reopening['Model_High'] = reopening['Model_High'].astype('str')
reopening['GradeLevels_Remote_Phasein'] = reopening['GradeLevels_Remote_Phasein'].astype('str')
reopening['StudentGroups_Remote_Phasein'] = reopening['StudentGroups_Remote_Phasein'].astype('str')
reopening['Other_Remote_Phasein'] = reopening['Other_Remote_Phasein'].astype('str')
```

```python
# reopening df - Check mixed data columns are fixed

for col in reopening.columns.tolist():
      weird = (reopening[[col]].applymap(type) != reopening[[col]].iloc[0].apply(type)).any(axis = 1)
      if len (reopening[weird]) > 0:
        print (col)
```

```python
# student df - Check for columns with mixed data types

for col in student.columns.tolist():
      weird2 = (student[[col]].applymap(type) != student[[col]].iloc[0].apply(type)).any(axis = 1)
      if len (student[weird2]) > 0:
        print (col)
```

## 03. Wrangle data

```python
# student df - drop columns that are not necessary for analysis - SchoolCode, SchoolName, SchoolOrganizationId, DataAsOf

student = student.drop(columns = ['SchoolCode'])
student = student.drop(columns = ['SchoolName'])
student = student.drop(columns = ['SchoolOrganizationId'])
student = student.drop(columns = ['DataAsOf'])
```

```python
# student df - check columns were removed

student.head(10)
```

```python
# reopening df - rename Week column for clarity

reopening.rename(columns = {'Week' : 'Week Of'}, inplace = True)
```

```python
# reopening df - check column was renamed

reopening.head(10)
```

```python
# reopening df - remove "Week of" text from all values in Week Of column

reopening['Week Of'] = reopening['Week Of'].str.replace(r'Week of ', '')
```

```python
# reopening df - check column values were updated

reopening['Week Of'].value_counts(dropna = False)
```

## 04. Merge datasets

```python
# Check shape of student dataset

student.shape
```

```python
# Check shape of reopening dataset

reopening.shape
```

```python
# Merge student df and reopening df using district name as key, add a merge flag

df_merged = reopening.merge(student, on = ['DistrictName'], indicator = True)
```
